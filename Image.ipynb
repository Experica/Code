# Choose Dataset

```julia
using Revise,NeuroAnalysis,Interact
droot = "../VLABDataExport";rroot = "../Result";iroot = "../NaturalStimuli";test = "[Image]"
sdir = dropdown(filter!(x->!startswith(x,"."),readdir(droot)),label="Subject")
fname = map(x->dropdown(getexpericafile(subject=x,test=test,dir="$droot/$x"),label="Experiment"),observe(sdir))
hbox(sdir,fname)
```

# Prepare Dataset

```julia
dataset = prepare("$droot/$(sdir[])/$(fname[][])","spike")
```

## Condition Test

```julia
using Statistics,Plots,Images
ex = dataset["ex"];envparam = ex["EnvParam"];preicidur = ex["PreICI"];conddur = ex["CondDur"];suficidur = ex["SufICI"]
bgcolor=RGBA(getparam(envparam,"BGColor")...)
imagesetname = replace(getparam(envparam,"ImageSet","ImageQuad"),"Ã—"=>"_")
imagemasktype = getparam(envparam,"MaskType","ImageQuad")
imagemaskradius = getparam(envparam,"MaskRadius","ImageQuad")
imagemasksigma = getparam(envparam,"Sigma","ImageQuad")
ct,ctc = ctctc(ex)
vcti = .!(isnan.(ct[:CondOn]) .& isnan.(ct[:CondOff]));ct=ct[vcti,:];ctc=ctc[vcti,:]
fl,fli,fln=flin(ctc)
condon = ct[:CondOn];condoff = ct[:CondOff];mconddur=mean(condoff-condon)
gr()
default(size=(650,420), leg=true)
histogram(condoff-condon,bins=20,title="Condition Duration(Set to $conddur ms)")
```

## Image Set

```julia
imageset = loadimageset("$iroot/$imagesetname",name=fl[:Image],alpha=true)
bgimagecolor = oftype(imageset[1][1],bgcolor)
display(rand(imageset,3,8))
```

## Prepare Image Stimuli

```julia
nscale = 2;downsample = 2;sigma = 1.5
imagestimuli = map(i->gaussian_pyramid(i, nscale-1, downsample, sigma),imageset)
imagesize = map(i->size(i),imagestimuli[1])
unmaskindex = map(i->alphamask(i,radius=imagemaskradius,sigma=imagemasksigma,masktype=imagemasktype)[2],imagestimuli[1])
imagestimuli = map(s->map(i->alphablend.(alphamask(i[s],radius=imagemaskradius,sigma=imagemasksigma,masktype=imagemasktype)[1],[bgimagecolor]),imagestimuli),1:nscale);
```

# Spiking Activity

```julia
spike = dataset["spike"];spikeeid = spike["electrodeid"];spikeuuid = spike["uuid"];spikeuid = spike["unitid"];spiketime = spike["time"]
eid = togglebuttons(spikeeid,label="Electrode")
stext = slider(0:1000,value=30,label="Spike Extention")
vbox(eid,stext)
```

```julia
eididx = findfirst(spikeeid.==eid[]);est = spiketime[eididx];esu = spikeuid[eididx]
sts,ns,ws,is = subrv(est,condon.-stext[],condoff.+stext[],isminzero=true)
sts = map(x->x.-stext[],sts)
plotspiketrain(sts,uids=map(i->esu[i],is),timeline=[0,mconddur],title="Electrode$(eid[])")
```

# Spike Triggered Average

```julia
scaleindex1 = slider(1:nscale,value=1,label="Scale")
```

```julia
display(rand(imagestimuli[scaleindex1[]],3,8))
```

```julia
ximagesize1 = imagesize[scaleindex1[]]
x1 = Array{Float64}(undef,size(ctc,1),prod(ximagesize1))
for i in 1:size(ctc,1)
    x1[i,:]=vec(gray.(imagestimuli[scaleindex1[]][findfirst(fl[:Image].==ctc[:Image][i])]))
end
uid = map(e->togglebuttons(["All";spikeuuid[findfirst(spikeeid.==e)]],label="Unit"),observe(eid))
```

```julia
@manipulate for delay=0:200,decor=[false,true]
    _,y1,_,_ = subrv(uid[][]=="All" ? est : est[esu.==uid[][]],condon.+delay,condoff.+delay,israte=false)
    plotsta(sta(x1,y1,ximagesize1,decor=decor),delay=delay,decor=decor)
end
```

# ePPR

```julia
scaleindex2 = slider(1:nscale,value=2,label="Scale")
```

```julia
display(rand(imagestimuli[scaleindex2[]],3,8))
```

```julia
pixelscale = 255
ximagesize2 = imagesize[scaleindex2[]]
xi = unmaskindex[scaleindex2[]]
x2 = Array{Float64}(undef,size(ctc,1),prod(ximagesize2))
for i in 1:size(ctc,1)
    x2[i,:]=vec(gray.(imagestimuli[scaleindex2[]][findfirst(fl[:Image].==ctc[:Image][i])]))*pixelscale
end
uid = map(e->togglebuttons(["All";spikeuuid[findfirst(spikeeid.==e)]],label="Unit"),observe(eid))
delay = slider(0:100,value=20,label="Delay")
vbox(uid,delay)
```

```julia
y2 = subrvr(uid[][]=="All" ? est : est[esu.==uid[][]],condon.+delay[],condoff.+delay[]);
```

```julia
using ePPR
i = slider(1:size(x2,1))
xvalue = map(j->colorview(Gray,reshape(x2[j,:]/pixelscale,ximagesize2)),observe(i))
yvalue = map(j->y2[j],observe(i))
display.([i,xvalue,yvalue]);
```

```julia
hp = ePPRHyperParams(ximagesize2...,xindex=xi,blankcolor=gray(bgimagecolor)*pixelscale,nft=[2,2,2],lambda=1000)
model,models = epprhypercv(x2,y2,hp,ePPRDebugOptions(level=DebugVisual));
```
