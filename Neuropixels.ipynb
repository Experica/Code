# Neuropixels 1.0 Probe

```julia
nelectrode = 960
nchannel = 384
ncolumn = 2
nrow = Int(nelectrode/ncolumn)
nrowch = Int(nchannel/ncolumn)
pitch = (20, 16)
electrodesize = (12,12)

shanksize = (10000, 70, 24)
tiplength = 175
tipangle = 20

electrodepositions = vcat([[i*pitch[1] (2j-1)*pitch[2] + (isodd(i) ? pitch[2] : 0)] for i in 1:nrow, j in 1:ncolumn]...);
```

```julia
using Plots
scatter(electrodepositions[:,2],electrodepositions[:,1],marker=:rect,markerstrokewidth=0,markersize=2,legend=false,size=(600,1000))
```
