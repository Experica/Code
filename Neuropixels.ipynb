# Neuropixels 1.0 Probe

```julia
using Plots

nelectrode = 960
nchannel = 384
ncolumn = 2
nrow = Int(nelectrode/ncolumn)
nrowch = Int(nchannel/ncolumn)
electrodesize = (12,12) # Î¼m
pitch = (20, 16)
refch = [192,576,960] # for bank 0, 1, 2

shanksize = (10000, 70, 24)
tiplength = 175
tipangle = 20

apfs = 30000
lffs = 2500
adcbit = 10
datarate = 20.475 # MB/s
```

### Electrodes Layout

```julia
electrodepositions = vcat(([j*2*pitch[2] + (isodd(i) ? 0 : pitch[2]) ;; i*pitch[1]] for j in 0:ncolumn-1, i in 0:nrow-1)...)
color = fill(:gray35,nelectrode);color[refch].=:red
scatter(electrodepositions[:,1],electrodepositions[:,2];marker=:rect,markerstrokewidth=0,markersize=3,
        ratio=:equal,frame=:none,legend=false,color,size=(4*15,nrow*10))
```

### Recording Channels Layout

```julia
channelpositions = vcat(([j*2*pitch[2] + (isodd(i) ? 0 : pitch[2]) ;; i*pitch[1]] for j in 0:ncolumn-1, i in 0:nrowch-1)...)
color = fill(:gray35,nchannel);color[refch[1]]=:red
scatter(channelpositions[:,1],channelpositions[:,2];marker=:rect,markerstrokewidth=0,markersize=3,
        ratio=:equal,frame=:none,legend=false,color,size=(4*15,nrowch*10))
```
