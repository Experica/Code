# Optical Imaging - Single Trial Response

## Choose Dataset

```julia
using Interact
droot = "../VLABDataExport";rroot = "../Result"
sdir = dropdown(filter!(x->!startswith(x,"."),readdir(droot)),label="Subject")
oifile = map(x->dropdown(readdir("$droot/$x"),label="OIBlock"),observe(sdir))
hbox(sdir,oifile)
```

## Analyze OI Response

```julia
nbaseframes = slider(0:2,value=0,label="nbaseframes")
```

```julia
using Revise, NeuroAnalysis
block = prepare("$droot/$(sdir[])/$(oifile[][])")
nframesperstim = block["imagehead"]["nframesperstim"]
baseframeindex = nbaseframes[]<=0 ? nothing : 1:nbaseframes[]
frameindex = nbaseframes[]<=0 ? nothing : nbaseframes[]+1:nframesperstim
response = oiframeresponse(block["image"],frameindex=frameindex,baseframeindex=baseframeindex);
```

## Response

```julia
using Colors,Images
resultmaps=Set()
mapname = :rmap
push!(resultmaps,mapname)
eval(:($mapname=Gray.(clampscale(response,4))))
```

## Color Codes

```julia
using DataStructures
fch = OrderedDict("mScarlet"=>10,"mCherry"=>0,"mNeongreen"=>100,"GFP"=>130)
display(map(i->HSV(i,1,1),values(fch)))
color = togglebuttons(fch)
```

```julia
mapname = :rcmap
push!(resultmaps,mapname)
eval(:($mapname=map(i->HSV(color[],1,i),clampscale(response,4))))
```

# Align Map

```julia
oireffile = map(x->dropdown(filter!(x->!endswith(x,"mat"),readdir("$droot/$x")),label="OIReference"),observe(sdir))
```

```julia
oiref = load("$droot/$(sdir[])/$(oireffile[][])")
```

```julia
alpha = slider(0:0.1:1,value=0.6,label="Map Alpha")
```

```julia
mapname = :oirefblendrmap
push!(resultmaps,mapname)
eval(:($mapname=alphablend.(RGB.(rmap),oiref,alpha[])))
```

```julia
mapname = :oirefblendrcmap
push!(resultmaps,mapname)
eval(:($mapname=alphablend.(RGB.(rcmap),oiref,alpha[])))
```

# Save Results

```julia
rdir = "$rroot/$(sdir[])"
rfhead = splitext(oifile[][])[1]
isdir(rdir) || mkdir(rdir)
eval.(:(save("$rdir/$(rfhead)_"*$(string(i))*".png",$i)) for i in resultmaps);
```
