# Read and Prepare Dataset

```julia
using NeuroAnalysis,Plots,Interact,DataFrames,Colors,Images,ImageFiltering,ProgressMeter
droot = "../VLABDataExport";rroot = "../Result"
sdir = dropdown(filter!(x->x[1]!='.',readdir(droot)),label="Subject")
vlabfile = map(x->dropdown(readdir("$droot/$x"),label="VLab"),signal(sdir))
oifile = map(x->dropdown(readdir("$droot/$x"),label="OIBlock"),signal(sdir))
map(display,[sdir,vlabfile,oifile]);
```

```julia
dataset = prepare("$droot/$(sdir.value)/$(vlabfile.value.value)");ex=dataset["ex"];cond = ex["Cond"]
oifiles = getoifile(subject=ex["Subject_ID"],experimentid=match(r"E([0-9]+)B",oifile.value.value).captures[1],dir=abspath("$droot/$(sdir.value)"))
image=DataFrame(source=oifiles)
dataset["image"]=image;
```

# Map Analysis

```julia
block = prepare(image[:source][1])
```

```julia
filter=nothing;stimuli = Int[];response=Any[]
p = ProgressMeter.Progress(size(image)[1],1,"Computing Response ...",50)
for i in image[:source]
    block = prepare(i)
    nframesperstim = block["imagehead"]["nframesperstim"]
    stimuli=cat(1,stimuli,block["imagehead"]["listofstimuli"])
    response = push!(response,oiframeresponse(block["image"],frameindex=nothing,filter=filter))

    if !haskey(dataset,"imagehead")
        dataset["imagehead"]=block["imagehead"]  
    end
    next!(p)
end
image[:stimuli] = stimuli;image[:response] = response;
```

```julia
blank,cocktail,oimap = oiresponse(response,stimuli,ustimuli=1:2:17,blankstimuli=17,sdfactor=3);
```

## Color Map

```julia
maptype = dropdown(["cocktailmap","blankmap","map"],label="MapType")
```

```julia
filter = Kernel.DoG((3,30))
cmap,amap,mmap,theta,fl = oicomplexmap(oimap[Symbol(maptype.value)],Int.((oimap[:stimuli]+1)/2),cond,filter=filter,sdfactor=3);
```

### Angle Map

```julia
anglemap=map(a->HSV(rad2deg(a),1,1),amap)
```

### Polar Map

```julia
polarmap=map((a,m)->HSV(rad2deg(a),1,m),amap,mmap)
```

### Stimuli Color Codes

```julia
ctheta=Complex.(cos.(theta),sin.(theta))
show(fl)
colorcodes=HSV.(rad2deg.(theta),1,1)
```

### Angle Domain Map

```julia
anglemapd=map(a->HSV(rad2deg(anglemode(a,theta,ctheta)),1,1),amap)
```

### Polar Domain Map

```julia
polarmapd=map((a,m)->HSV(rad2deg(anglemode(a,theta,ctheta)),1,m),amap,mmap)
```
