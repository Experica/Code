# Prepare Metadata

```julia
using NeuroAnalysis
includet("Batch.jl")
env = Dict{Any,Any}(
    :dataroot => "../Data",
    :dataexportroot => "../DataExport",
    :resultroot => "../Result",
    :imageroot => "../NaturalStimuli")
meta = readmeta("$(env[:dataexportroot])/metadata.mat")
```

# Query Tests

```julia
testid = "OriSFColor"
tests=querymeta(meta,test=testid,sourceformat="SpikeGLX",subject="AE9")
```

# Flash of Two Colors

```julia
param = copy(env)
ur,uc = batchtests(tests[1:4,:],param,isplot=true);
```

# Hartley Subspace Parametric and Image Response 

```julia
param = copy(env)
param[:model]=[:STA]
param[:epprndelay]=1
param[:epprnft]=[2]
param[:epprlambda]=10
ur,uc = batchtests(tests[1:4,:],param,isplot=true);
```

# Drifting Grating with Ori and SpatialFreq

```julia
param = copy(env)
param[:responsedelay] = 0.02
param[:blank] = (:Ori, "blank")
param[:corrcond] = :allcond
ur,uc = batchtests(tests[1:3,:],param,isplot=true);
```

# Drifting Grating with Ori, SpatialFreq and Color

```julia
param = copy(env)
param[:responsedelay] = 0.02
param[:blank] = (:ColorID, 36)
param[:corrcond] = :allcond
ur,uc = batchtests(tests[1:1,:],param,isplot=true);
```
